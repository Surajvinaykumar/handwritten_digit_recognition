```python
import pygame
import sys
from pygame.locals import *
import numpy as np
from keras.models import load_model
import cv2

# Constants
WINDOWSIZEX = 1000
WINDOWSIZEY = 1000

BOUNDARYINC = 10
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)

IMAGESAVE = False

# Load model and labels
MODEL = load_model("bestmodel.keras")
LABELS = {0: "Zero", 1: "One", 2: "Two",
          3: "Three", 4: "Four", 5: "Five",
          6: "Six", 7: "Seven", 8: "Eight", 9: "Nine"}

# Initialize pygame
pygame.init()

# Set up display
DISPLAYSURF = pygame.display.set_mode((WINDOWSIZEX, WINDOWSIZEY))

pygame.display.set_caption("Digit Board")

# Initialize variables
iswriting = False
number_xcord = []
number_ycord = []

image_count = 1
PREDICT = True
FONT = pygame.font.Font(None, 36)  # Default font

while True:
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()

        elif event.type == MOUSEMOTION and iswriting:
            xcord, ycord = event.pos
            pygame.draw.circle(DISPLAYSURF, WHITE, (xcord, ycord), 4,0)
            number_xcord.append(xcord)
            number_ycord.append(ycord)

        elif event.type == MOUSEBUTTONDOWN:
            iswriting = True

        elif event.type == MOUSEBUTTONUP:
            iswriting = False
            number_xcord = sorted(number_xcord)
            number_ycord = sorted(number_ycord)

            rect_min_x, rect_max_x = max(number_xcord[0] - BOUNDARYINC, 0), min(WINDOWSIZEX, number_xcord[-1] + BOUNDARYINC)
            rect_min_y, rect_max_y = max(number_ycord[0] - BOUNDARYINC, 0), min(WINDOWSIZEY, number_ycord[-1] + BOUNDARYINC)

            number_xcord = []
            number_ycord = []

            img_arr = pygame.surfarray.array3d(DISPLAYSURF)[rect_min_x:rect_max_x, rect_min_y:rect_max_y].astype(np.float32)
            img_arr = cv2.cvtColor(img_arr, cv2.COLOR_RGB2GRAY)
            img_arr = cv2.resize(img_arr, (8,8))
            img_arr = np.pad(img_arr, ((10, 10), (10, 10)), mode='constant', constant_values=0)
            img_arr = img_arr.reshape(1, 28, 28, 1).astype(np.float32) / 255.0

            if IMAGESAVE:
                cv2.imwrite(f"image_{image_count}.png", img_arr * 255)
                #image_count += 1

            if PREDICT:
                prediction = MODEL.predict(img_arr)
                predicted_label = np.argmax(prediction)
                label_text = LABELS[predicted_label]

                text_surface = FONT.render(label_text, True, RED)
                text_rect  = text_surface.get_rect()
                text_rect.center = (rect_min_x + 50, rect_max_y + 20)

                DISPLAYSURF.blit(text_surface, text_rect)

        elif event.type == KEYDOWN:
            if event.unicode == "n":
                DISPLAYSURF.fill(BLACK)

    pygame.display.update()
```
